```python
def manage_list():
  items = []  # empty shopping cart list

  while True:  # loop until user enters quit
    print("Enter 'a' to add an item, 'd' to delete an item, 'v' to view the current list, or 'q' to quit.")
    action = input()  # empty input

    if action == 'a': 
      item = input("Enter the item you want to add: ")
      items.append(item)
    elif action == 'd': 
      item = input("Enter the item you want to delete: ") 
      if item in items: 
        items.remove(item) 
        print("Item deleted.")
      else: 
        print("Item not found in the list.")
    elif action == 'v':  
      print("Here is the current list:")
      for item in items:  
        print(item) 
    elif action == 'q': 
      break  
    else:   
      print("This action can't be taken. Please re-enter your query with the approved actions list.")

  # print out final list
  print("Final list of items:")
  for item in items:
    print(item)

manage_list()
```
